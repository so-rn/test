```python
import pandas as pd

dataset = pd.read_csv(r"/home/soraan/Desktop/karzanmahmoudi/03_Visualization/day3/internet acces/internet.csv")

dataset.columns = ["Country", "Code", "Year", "Internet_Usage"]
print(dataset.head(5))
```

```python
dataset_filtered = dataset[dataset["Year"] >= 2000]

print(dataset_filtered.head(5))
```

```python
import plotly.express as px
import json

with open(r"/home/soraan/Desktop/karzanmahmoudi/03_Visualization/day3/internet acces/countries.geojson") as file:
    map_data = json.load(file)

world_map = px.choropleth(
    dataset_filtered,
    geojson=map_data,
    locations="Code",
    featureidkey="properties.ISO_A3",
    color="Internet_Usage",
    hover_name="Country",
    animation_frame="Year",
    color_continuous_scale="Blues",
    range_color=[0, 100],
    projection="natural earth"
)

world_map.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})

world_map.show()
```

```python
countries_to_compare = ["Switzerland", "Germany", "United States" , "Iran"]
dataset_compare = dataset_filtered[dataset_filtered["Country"].isin(countries_to_compare)]

line_chart = px.line(
    dataset_compare,
    x="Year",
    y="Internet_Usage",
    color="Country",
    title = "Compare Internet USA , Iran , Germany and Switzerland",
    markers=True
)

line_chart.show()
```
