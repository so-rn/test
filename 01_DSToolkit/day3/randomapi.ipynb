```python
import requests

url = "https://transport.opendata.ch/v1/connections"

params = {
    "from": "Baden",
    "to": "Zurich",
    "limit": 2 
}

response = requests.get(url, params=params)
data = response.json()

if response.status_code == 200:
    print(f"--- Next trains from Baden to Zurich ---")
    for connection in data['connections']:
        departure_time = connection['from']['departure']
        platform = connection['from']['platform']
        duration = connection['duration']
        
        print(f"Departure: {departure_time}")
        print(f"Platform: {platform}")
        print(f"Duration: {duration}")
        print("-" * 30)
else:
    print(f"Error fetching transport data: {response.status_code}")
```
