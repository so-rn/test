```python
import pandas as pd

df = pd.read_csv(r'C:\Users\SoRaaN\Desktop\Day2\IGN\ign.csv')
# mige avalin radifo vardsr 
sorted_df = df.sort_values(by=['release_year', 'release_month'])

first_game = sorted_df.iloc[0]
print("1. The first game is:")

print(f"Title: {first_game['title']} | Year: {first_game['release_year']} | Month: {first_game['release_month']}")

#inja bayad hameye baziha ro be onvane yek group begirim va score ha ro be onvane miangin hesab konim
yearly_variation = df.groupby('release_year')['score'].mean()

# inja chapesh mikonim 
print("\n2. Yearly variation of average scores:")
print(yearly_variation)

# inja vase hame platformaro goroh bandi va minagin emtizat ro migire  ro migire 
platform_avg = df.groupby('platform')['score'].mean()


#inja check mikone bozorgtarinesho va esmesham check mikone va behtarinesho entrekhab mikoneo chap mishe
best_platform = platform_avg.idxmax()
print(f"\n3. The best platform is: {best_platform}")
print(f"Average Score: {platform_avg.max():.2f}")

# borto soton score farse ro check kon 
# harja amazin o masterpiece ro didi beriz to blockbuster
blockbusters = df[df['score_phrase'].isin(['Amazing', 'Masterpiece'])]

#inja mikhaim bebibim bvaziha che mahi montasher mishe
#dastor valuecont mige az in list masalan ye tedsadish to flan mah bocde
#inja migim sort_index ke be tartib mah ha ro chap kone masalan 1 ta 12
monthly_blockbusters = blockbusters['release_month'].value_counts().sort_index()

#injam migim in chza ba in moshakhaste inja chap mishe
print("\n4. Release timing of 'Amazing' and 'Masterpiece' games (by month):")
print(monthly_blockbusters)

```
