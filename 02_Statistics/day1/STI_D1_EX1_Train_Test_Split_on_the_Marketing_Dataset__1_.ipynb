````python
import pandas as pd
import matplotlib.pyplot as plt

data_file='https://drive.google.com/uc?export=download&id=1-6CGpOFdmZ8vOz3OrCXZC170q1zJDJtQ'
df = pd.read_csv(data_file)
df
````

````python
plt.figure(figsize=(6, 4))
df['conversion'].hist(bins=[-0.5, 0.5, 1.5], rwidth=0.8)
plt.xticks([0, 1])
plt.title('Total Conversion Distribution')
plt.show()
````

````python
# raveshe awal random hesab konim
train_rand = df.sample(frac=0.75, random_state=100)
test_rand = df.drop(train_rand.index)

#raveshe dowal tsartibitrain_
train_seq = df.head(int(0.75 * len(df)))
test_seq = df.drop(train_seq.index)
````

````python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# raveshe tasadfi nemodar
axes[0].hist(train_rand['conversion'], alpha=0.5, label='Train', bins=[-0.5, 0.5, 1.5], rwidth=0.7)
axes[0].hist(test_rand['conversion'], alpha=0.5, label='Test', bins=[-0.5, 0.5, 1.5], rwidth=0.5)
axes[0].set_title('Random Split')
axes[0].legend()

# raveshe tsartibi nemodar
axes[1].hist(train_seq['conversion'], alpha=0.5, label='Train', color='green', bins=[-0.5, 0.5, 1.5], rwidth=0.7)
axes[1].hist(test_seq['conversion'], alpha=0.5, label='Test', color='red', bins=[-0.5, 0.5, 1.5], rwidth=0.5)
axes[1].set_title('Sequential Split')
axes[1].legend()

plt.show()
````

````python
# Check the mean of 'history' for both methods
print("random Mean:", test_rand['history'].mean())
print("tartibi Mean:", test_seq['history'].mean())
````
