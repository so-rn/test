## Exercise 1
Use the titanic.csv dataset to answer the following questions:

- Are passenger gender and cabin class related?
- Does the age of passengers vary across cabin class?


````python
import pandas as pd
import scipy.stats as stats


path = r"C:\Users\SoRaaN\Desktop\karzanmahmoudi\02_Statistics\day3\titanic.csv"

df = pd.read_csv(path)
print(df.columns)
````

````python
# reading data from csv

ct = pd.crosstab(df['Sex'], df['PClass'])
print("Gender vs Class Table:")
print(ct)

chi2, p_val, dof, expected = stats.chi2_contingency(ct)
print(f"\nChi-square P-value: {p_val:.4e}")
````

### Titanic PT 1: Are passenger gender and cabin class related?

````python
# your code here

contingency_table = pd.crosstab(df['Sex'], df['PClass'])
print("Contingency Table:")
print(contingency_table)

chi2, p_val, dof, expected = stats.chi2_contingency(contingency_table)

print(f"\nChi-square Statistic: {chi2:.4f}")
print(f"P-value: {p_val:.4e}")

````

### Titanic PT 2: Does the age of passengers vary across cabin class?

````python
# your code here
c1 = df[df['PClass'] == '1st']['Age'].dropna()
c2 = df[df['PClass'] == '2nd']['Age'].dropna()
c3 = df[df['PClass'] == '3rd']['Age'].dropna()


f_stat, p_val_anova = stats.f_oneway(c1, c2, c3)
print(f"ANOVA P-value: {p_val_anova:.4e}")
````

## Plant Growth Dataset

### Plant Growth PT 1: Did the treatments have an effect?

````python
# reading data from csv
import pandas as pd
data_path = r"C:\Users\SoRaaN\Desktop\karzanmahmoudi\02_Statistics\day3\plant_growth.csv"
df_plant = pd.read_csv(data_path)
df_plant
````

````python
# your code here
ctrl = df_plant[df_plant['group'] == 'ctrl']['weight'].dropna()
trt1 = df_plant[df_plant['group'] == 'trt1']['weight'].dropna()
trt2 = df_plant[df_plant['group'] == 'trt2']['weight'].dropna()

f_stat, p_val_anova = stats.f_oneway(ctrl, trt1, trt2)

print(f"ANOVA F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val_anova:.4f}")
````

## Exercise 3

Using the titanic.csv again, answer the following questions:

- Were the passengers who survived older or younger than those who did not?
- Were men older than women on average?


````python
male_age = df[df['Sex'] == 'male']['Age'].dropna()
female_age = df[df['Sex'] == 'female']['Age'].dropna()

t_stat_gender, p_val_gender = stats.ttest_ind(male_age, female_age, equal_var=False)

print(f"Gender vs Age - T-statistic: {t_stat_gender:.4f}")
print(f"Gender vs Age - P-value: {p_val_gender:.4e}")

print(f"Mean age of men: {male_age.mean():.2f}")
print(f"Mean age of women: {female_age.mean():.2f}")
````

### PT 1: Were the passengers who survived older or younger than those who did not?


````python
# your code here

import scipy.stats as stats

survived = df[df['Survived'] == 1]['Age'].dropna()
not_survived = df[df['Survived'] == 0]['Age'].dropna()

t_stat_surv, p_val_surv = stats.ttest_ind(survived, not_survived, equal_var=False)

print(f"Survival vs Age - T-statistic: {t_stat_surv:.4f}")
print(f"Survival vs Age - P-value: {p_val_surv:.4e}")
````
