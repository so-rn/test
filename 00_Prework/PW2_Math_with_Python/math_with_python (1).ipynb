## Exercise 9

Find the critical points of the function $f(x)= x^3−2x^2+x$ and use the second derivative test to find the maximum on interval $x⊂[0,1]$. 

**Hint 1**: remember the concepts of first differentiation and second differentiation;

**Hint 2**: `sympy` python library with use of `solve`, `diff`.

```python
import sympy as sp

x = sp.Symbol('x')

f = x**3 - 2*x**2 + x

f_prime = sp.diff(f, x)

critical_points = sp.solve(f_prime, x)

print(f"Critical Points: {critical_points}")

f_double_prime = sp.diff(f_prime, x)

for point in critical_points:
    if 0 <= point <= 1:
        test_value = f_double_prime.subs(x, point)
        if test_value < 0:
            print(f"Point {point} is a Maximum (double prime = {test_value})")
        elif test_value > 0:
            print(f"Point {point} is a Minimum (double prime = {test_value})")
```

## Exercise 10

Maximise $f(x, y, z) = xy + yz$, subject to the constraints $x + 2y = 6$ and $x − 3z = 0$. 

**Hint 1**: set up the Lagrange equation, $f (x, y, z) − λ(constraint1) − μ(constraint2)$; 

**Hint 2**: Set partial derivatives to zero and solve the following set of equations; 

**Hint 3**: Use `solve`. 

```python
import sympy as sp

x, y, z, lam, mu = sp.symbols('x y z lam mu')

f = x*y + y*z
g1 = x + 2*y - 6
g2 = x - 3*z

L = f - lam * g1 - mu * g2

eqs = [
    sp.diff(L, x),
    sp.diff(L, y),
    sp.diff(L, z),
    sp.diff(L, lam),
    sp.diff(L, mu)
]

solution = sp.solve(eqs, (x, y, z, lam, mu))

print(solution)
```
